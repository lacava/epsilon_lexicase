```python
# examing the effect of population size and test case size on probability of selection.

# probabilitiy of a case coming first at least once

import numpy as np
import itertools as it
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

%matplotlib inline
plt.rcParams['figure.figsize'] = (6, 4)
plt.rcParams['xtick.labelsize'] = 14

N = np.arange(2,10000,10)
T = np.arange(2,10000,10)

N,T = np.meshgrid(N,T)

Pcf = 1 - ((T-1)/T)**N
        
    
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax=plt.contourf(N,T,Pcf, 100, cmap = cm.coolwarm, antialiased=True)
plt.xlabel('\nN',size=16)
plt.xticks(size=12)
plt.ylabel('\nT',size=16)
plt.yticks(size=12)

# Add a color bar which maps values to colors.
cb = fig.colorbar(ax, shrink=1, aspect=7)
cb.set_label(label="$P_{first}$", size=16)
cb.set_ticks(np.arange(0,1.1,step=0.1))
fig.tight_layout()
fig.savefig('../../../Papers/epsilon_lexicase_journal/figs/prob_case_first_heatmap.pdf',figsize=(6,4))
fig.savefig('../../../Papers/epsilon_lexicase_journal/figs/prob_case_first_heatmap.png')
plt.show()
```
